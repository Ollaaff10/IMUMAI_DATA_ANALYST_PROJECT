```python
# Importing necessary libraries
#loading and exploration of data

import pandas as pd

# File path and loading datasets
car_golf_dataset = pd.read_csv(r'D:\Salman work folder\ImumAi_Data_Analyst_Test\car-golf-dataset.csv', delimiter=';')
car_golf_info = pd.read_csv(r'D:\Salman work folder\ImumAi_Data_Analyst_Test\car-golf-info.csv', delimiter=';')

# Displaying basic information about the datasets
print("Car Golf Dataset Info:")
print(car_golf_dataset.info())
print("\nCar Golf Info Dataset Info:")
print(car_golf_info.info())

# Previewing the datasets
print("\nCar Golf Dataset Sample:")
print(car_golf_dataset.head())
print("\nCar Golf Info Sample:")
print(car_golf_info.head())
```

```python
# Checking for missing values in both datasets
print("\nMissing Values in Car Golf Dataset:")
print(car_golf_dataset.isnull().sum())

print("\nMissing Values in Car Golf Info:")
print(car_golf_info.isnull().sum())
```

```python
# Geting unique values in important columns to understand the variability
print("\nUnique Categories in Car Golf Dataset:")
print(car_golf_dataset['category'].unique())

print("\nUnique Fuel Types in Car Golf Dataset:")
print(car_golf_dataset['fuel'].unique())

print("\nUnique Transmission Types in Car Golf Dataset:")
print(car_golf_dataset['transmission'].unique())

print("\nUnique Parsed Categories in Car Golf Info:")
print(car_golf_info['parsed_category'].unique())

print("\nUnique Parsed Engine Types in Car Golf Info:")
print(car_golf_info['parsed_engine_type'].unique())

```

```python
# Checking for mismatched or contradictory data
# Example: Cars with "Manual gearbox" but DSG in the title
suspicious_transmission = car_golf_dataset[
    (car_golf_dataset['transmission'].str.contains("Manual", na=False)) &
    (car_golf_dataset['title'].str.contains("DSG", na=False))
]
print("\nSuspicious Transmission Rows:")
print(suspicious_transmission)

# Example: Cars with "Petrol" fuel but mention "Hybrid" in the title
suspicious_fuel = car_golf_dataset[
    (car_golf_dataset['fuel'].str.contains("Petrol", na=False)) &
    (car_golf_dataset['title'].str.contains("Hybrid", na=False))
]
print("\nSuspicious Fuel Rows:")
print(suspicious_fuel)
```
